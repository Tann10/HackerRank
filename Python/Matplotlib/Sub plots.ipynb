```python
import pandas as pd

import matplotlib.pyplot as plt

plt.style.use('seaborn')
```

Figure is the container holding plots, graph shown by the code
Axes are the actual plots, a figure can have multiple plots 

```python
data = pd.read_csv('filling-data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# instantiate figure and axes - default 1 by 1 
# pass sharex to set share axis
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')


ax1.legend()

ax2.legend()

# in case of OOP use "set_" before naming title and axes
ax1.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.set_ylabel('Median Salary (USD)')


plt.show()
```
