```python
x_vals = []
y_vals = []

index=count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    
    plt.cla()
    plt.plot(x_vals, y_vals)
    

    
# call funcanimation from matplotlib with parameter as 
#current figure,function, interval in secs, 1sec = 1000
ani = FuncAnimation(plt.gcf(), animate, interval=1000)    

plt.tight_layout()
plt.show()
```
